```python
# Package Imports

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy import stats
# import requests

```

```python
# Read CSV

df = pd.read_csv('chicagoaggdata-chicagatrafficdata-QueryResult.csv')

print(df)
```

```python
# Vehicle - Type filter to only non-commetcial vehicles

gp_vehicle_type = df.groupby(['vehicle_type']).size()
print(gp_vehicle_type)
```

```python
# Create data frame for column value passenger, sport utility vehicle, SUV, Van/Mini Van or pickup in vehicl etype
filter_vehicle_type = df[(df['vehicle_type'] == 'PICKUP') | (df['vehicle_type'] == 'PASSENGER') | (df['vehicle_type'] == 'SPORT UTILITY VEHICLE (SUV)') | (df['vehicle_type'] == 'VAN/MINI-VAN') ] 

print(filter_vehicle_type.groupby(['vehicle_type']).size())
filter_vehicle_type


```

```python
# Take rows from vehicle filter subset dataframe that does not have NaN values for make, model, and year 

filter_vt = filter_vehicle_type[df['model'].notna() & df['make'].notna() & df['vehicle_year'].notna()]
filter_vt 

```

```python
for col in filter_vt.columns:
    print(col)
```

```python
# Explore groupings to remove
gp_most_severe_injury = filter_vt.groupby(['most_severe_injury']).size()
print("Most Severe Injury Grouping", gp_most_severe_injury)
print('\n')
gp_injury_class = filter_vt.groupby(['injury_classification']).size()
print("Injury Classification", gp_injury_class)
print('\n')
gp_injury_total = filter_vt.groupby(['injuries_total']).size()
print("Injury Total", gp_injury_total)
print('\n')
gp_injuries_incapacitating = filter_vt.groupby(['injuries_incapacitating']).size()
print("Injury Incapacitating", gp_injuries_incapacitating)

```

```python
# Take rows from vehicle filter subset dataframe that does not have NaN values for crash severity
filter_injury_severity = filter_vt[df['most_severe_injury'].notna() & df['injury_classification'].notna() & df['injuries_total'].notna()]
filter_injury_severity 
```

```python
# Create subset where all roads had clear weather conditions 
gp_weather_cond = filter_vt.groupby(['weather_condition']).size()
print(gp_weather_cond)
filter_vt_weather = filter_vt[(filter_vt['weather_condition'] == 'CLEAR')]
filter_vt_weather
```

```python
# Create subset where all drivers were in normal condition
gp_physical_condition = filter_vt_weather.groupby(['physical_condition']).size()
# print(gp_physical_condition)
filter_vt_normal = filter_vt_weather[
    (filter_vt_weather['physical_condition'].isna()) | 
    (filter_vt_weather['physical_condition'] == 'NORMAL')
    ]

filter_vt_normal['physical_condition']
```

```python
# Create subset where all roads had good lighting conditions
gp_lighting_condition = filter_vt_normal.groupby(['lighting_condition']).size()
print(gp_lighting_condition)
filter_vt_lighting = filter_vt_normal[(filter_vt_normal['lighting_condition'] == 'DAYLIGHT')| (filter_vt_normal['lighting_condition'] == 'DARKNESS, LIGHTED ROAD') ]
filter_vt_lighting
```

```python
# Drop all rows where driver was driving in poor lighting conditions
gp_roadway = filter_vt_lighting.groupby(['roadway_surface_cond']).size()
print(gp_roadway)
filter_vt_roadway= filter_vt_lighting[(filter_vt_lighting['roadway_surface_cond'] == 'DRY')]
filter_vt_roadway
```

```python
# Drop all rows where driver was not wearing a seatbelt
gp_seatbelt = filter_vt_roadway.groupby(['safety_equipment']).size()
print(gp_seatbelt)
filter_vt_seatbelt= filter_vt_roadway[(filter_vt_roadway['safety_equipment'] == 'SAFETY BELT USED')]
filter_vt_seatbelt
```

```python
# Filter out roads that have no defects
gp_road_defect = filter_vt_seatbelt.groupby(['road_defect']).size()
print(gp_road_defect)
filter_vt_road_defect= filter_vt_seatbelt[(filter_vt_seatbelt['road_defect'] == 'NO DEFECTS')]
filter_vt_road_defect
```

```python
# Filter out so data so that only reports with driver's and passengers are taken into account
gp_person_type = filter_vt_road_defect.groupby(['person_type']).size()
print(gp_person_type)
df_people= filter_vt_road_defect[
    (filter_vt_road_defect['person_type'] == 'DRIVER') |
    (filter_vt_road_defect['person_type'] == 'PASSENGER') 
    ]
df_people
```

```python
# Filter out 'UNKNOWN' values for make, model and year 
df_final = df_people[
    (df_people['make'] != 'UNKNOWN') &
    (df_people['model'] != 'UNKNOWN')&
    (df_people['vehicle_year'] != 'UNKNOWN') &
    (df_people['vehicle_year'] > 1950) &
    (df_people['vehicle_year'] < 2021 )
    ]
df_final
```

```python
# Group By Make to select the most popular car makes in the data set

gp_make= df_final.groupby(['make']).size()
print("Filtered Dataset",gp_make)
print('\n')
gp_make_original= df.groupby(['make']).size()
print("Orginal Dataset", gp_make_original)

```

```python
# Group By Make to select the most popular car makes in the data set

gp_model = df_final.groupby(['model']).size()
print("Filtered Dataset",gp_model)
print('\n')
gp_model_original= df.groupby(['model']).size()
print("Orginal Dataset", gp_model_original)
```
